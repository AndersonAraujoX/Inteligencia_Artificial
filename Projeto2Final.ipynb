## Tecnicas Utilizadas:
### Para dados faltantes:
<p>
    Como o numero de dados faltantes por muito pequeno (menos de 1 por dataset), decidimos por eliminar as linhas contendo os nulos.
</p>

### Tratar dados não numericos:

+ String:
    + Como saida, tratamos com o LabelEncoder, ele transforma as strings em saidas numericas;
    + Como entrada, tratamos com o OneHotEncoder, transformando os dados em colunas, marcando 1 caso presente e 0 caso contrario;
+ Time:
    + transformamos dados Time em segundos, depois tratamos eles com algum Pré Processamento.
    
### Dados númericos

+ Dependendo do caso, utilizamos MinMaxEncoder() ou MaxAbsEnconder().


```python
df_abacate=df_abacate.dropna()
df_abacate.rename(columns = {'Unnamed: 0':'index'}, inplace = True)
assert pd.notnull(df_abacate).all().all()

from datetime import datetime, timedelta

def posix_time(dt):
    return (dt - datetime(1970, 1, 1)) / timedelta(seconds=1)


df_abacate['Date']=pd.to_datetime(df_abacate['Date']).apply(posix_time)

X = df_abacate.drop(['type'], axis=1)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_abacate=le.fit_transform(df_abacate['type'])

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
X_abacate = ohe.fit_transform(X)

from sklearn.preprocessing import MaxAbsScaler
scaler=MaxAbsScaler()
scaler_data=scaler.fit_transform(X_abacate)
x_abacate=scaler_data

```
